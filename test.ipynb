```python
import pandas as pd
import requests

BASE_URL = "https://mapprod3.environment.nsw.gov.au/arcgis/rest/services/EDP/Administrative_Boundaries/MapServer"

def get_all_sa4_info():
    url = f"{BASE_URL}/6/query"
    params = {
        "where": "STATE_NAME = 'New South Wales'",
        "outFields": "*",  # 获取所有列
        "returnGeometry": "false",
        "f": "json",
    }
    data = requests.get(url, params=params).json()
    rows = [feature["attributes"] for feature in data["features"]]
    return pd.DataFrame(rows)

# 获取完整的数据表格
df = get_all_sa4_info()

# === 打印查看的三种招式 ===

# 第 1 招：打印所有的“列名”（强烈推荐）
# 这能让你一眼看清数据库里到底存了哪些字段
print("【所有的列名】:")
print(df.columns.tolist())
print("-" * 50)

# 第 2 招：打印前 5 行数据
# 表格太宽太长时，只看前几行就能知道数据长什么样
print("【前 5 行数据预览】:")
print(df.head())
print("-" * 50)

# 第 3 招：如果你想强行看所有的行和列
# （需要先修改 Pandas 的显示设置，否则中间的列会被省略号替代）
pd.set_option('display.max_columns', None) # 显示所有列
print("【完整表格】:")
print(df)
```
